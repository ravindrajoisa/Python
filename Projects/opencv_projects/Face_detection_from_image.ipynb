```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = plt.imread("./group-photo.jpg")
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
detector = cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')    # here you are loading the model
```

```python
all_faces = detector.detectMultiScale(img, 1.2, 5)  # here 1.2 is the scaleFactor is used to improve accuracy of the model, minNeighbor, 5- how many rectangles should be retained
# without saleFactor ad minNeighbor, there are chances of additional rectangles being marked. This is done to improve accuracy of the model.
```

```python
all_faces.shape  # the output (4,4) indicates that it has identified 4 faces with 4 properties - x,y,w,h in each face - x,y are the coordinates, w-width, h-height
```

```python
all_faces
```

```python
x,y,w,h = all_faces[1]
```

```python
img_copy = img.copy()
```

```python
img = cv2.rectangle(img_copy, (x,y), (x+w, y+h), (255,0,0), 2)
# here is the syntax rectangle(image, top-left rect coordianates, bottom-right coordinates, color, rectangle thickness
```

```python
plt.imshow(img)
```

```python
# plotting rectangles for all faces
for face in all_faces:
    x,y,w,h = face
    img = cv2.rectangle(img_copy, (x,y), (x+w, y+h), (255,0,0), 2)
```

```python
plt.imshow(img)
```
