```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
# plot the distrbution of the data --> like a histogram
sns.distplot(iris['petal_length'], bins=20, kde=False)
plt.show()
```

```python
# the sign wave is called KDE - Kernel density estimation
# distplot is depricated, so use histplot
# you can set kde = 0 to remove the sign wave
```

```python
sns.kdeplot(iris['petal_length'])
plt.show()
```

```python
sns.scatterplot(x=iris['petal_length'], y = iris['petal_width'])
plt.show()
```

```python
sns.scatterplot(x='petal_length', y='petal_width', data=iris)
plt.show()
```

```python
iris
```

```python
sns.scatterplot(x='petal_length', y='petal_width', data=iris, hue='species')
plt.show()
```

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=iris)
plt.show()
```

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='hex')
plt.show()
```

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='reg')
plt.show()
```

```python
### Pair Plot
```

```python
sns.pairplot(data=iris)
```

```python
sns.pairplot(data=iris, hue='species')
```

```python
# Categorical Plot
```

```python
sns.countplot(x='species', data=iris)
plt.show()
```

```python
# box plot
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(x = 'species', y = 'petal_length', data = iris)
plt.show()
```

```python
# the horizontal line in the box above shows the mean value
```

```python
#using sepal instead of petal

plt.figure(figsize=(10,10))
sns.boxplot(x = 'species', y = 'sepal_length', data = iris)
plt.show()
```

```python
# better than boxplot is the violinplot, this shows the distribution.
sns.violinplot(x = 'species', y = 'petal_length', data = iris)
plt.show()
```

```python
# finding the correlation
```

```python
iris
```

```python
# make use of the correlation, we will have to ignore the last column as it has text and the value isn't numerical
```

```python
# Drop the 'species' column and calculate the correlation matrix. If the species column isn't dropped it will throw an error saying convert string to float.

correlation_matrix = iris.drop(columns=['species']).corr()
```

```python
correlation_matrix
```

```python
sns.scatterplot(x='sepal_length', y='sepal_length', data = iris)
```

```python
# one straight line +ve and perfect correlation, 1 means perfect correlation.
```

```python
sns.heatmap(correlation_matrix.corr())
plt.show()
```

```python
print(plt.colormaps())
```

```python
sns.heatmap(correlation_matrix.corr(), cmap='Blues')
plt.show()
```

```python
sns.heatmap(correlation_matrix.corr(), cmap='Blues', annot=True)
plt.show()
```
