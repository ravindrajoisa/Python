```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.shape
```

```python
tips.head()
```

```python
sns.countplot(x=tips['sex'], hue='sex', data=tips)
plt.show()
```

```python
tips['day'].unique()
```

```python
# out of all the days which day gets the highest amount
sns.barplot(x='day', y='total_bill', hue='day',  data=tips)
plt.show()
```

```python
# out of all the days which day gets the highest amount
sns.barplot(x='day', y='total_bill', hue='sex',  data=tips, estimator=np.mean)
plt.show()
```

```python
  # what do people prefer - going to lunch or dinner and on which day?
sns.barplot(x='time', y='total_bill', data = tips, hue='day')
plt.show()
```

```python
# do smokers give more tip?
sns.barplot(x='smoker', y='tip', data = tips)
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
plt.show()
```
