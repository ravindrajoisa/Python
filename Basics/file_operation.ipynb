```python
### File Operation - Read and Write Files in Python
```

```python
## Read a file
with open('example.txt', 'r') as file:
    content = file.read()
    print(content)  
```

```python
## Read a file line by line
with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip())  # strip() removes leading/trailing whitespace
```

```python
## Writing to a file
with open('example.txt', 'w') as file:
    file.write("Hello, World!\n")
    file.write("This is a test file.\n") # the existing content will be overwritten

## Appending to a file
with open('example.txt', 'a') as file:
    file.write("This line is appended.\n")
```

```python
#writing a list of lines to a file
lines = ["First line\n", "Second line\n", "Third line\n"]
with open('example.txt', 'a') as file:
    file.writelines(lines)
```

```python
# Binary files
with open('example.bin', 'wb') as file:
    data = b'\x00\xFF\x7A\x3C'
    file.write(data)  # Writing binary data
```

```python
# reading a binary file
with open('example.bin', 'rb') as file:
    binary_content = file.read()
    print(binary_content)  # Output: b'\x00\xffz<'  
```

```python
# read the content from a source file and write to a destination file.
# copying a text file
with open('example.txt', 'r') as src_file:
    content = src_file.read()
with open('destination.txt', 'w') as dest_file:
    dest_file.write(content) # Copy content to destination file by creating a new destination file, if the file isn't present.
```

```python
# writing and then reading the same file
with open('example.txt', 'w+') as file:
    file.write("Temporary file content.\n")
    file.seek(0)  # Move the cursor back to the beginning of the file
    temp_content = file.read()
    print(temp_content)  # Output: Temporary file content.
```
