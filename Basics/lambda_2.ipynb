```python
# Lambda functions in Python
# Lambda functions are small anonymous functions defined using the lambda keyword.
# They can take any number of arguments but can only have one expression.
# They are often used for short, throwaway functions that are not reused elsewhere in the code.

#Syntax:
# lambda arguments: expression 
```

```python
# example
add = lambda x, y: x + y
print(add(5, 3))  # Output: 8
```

```python
even = lambda x: x % 2 == 0
print(even(4))  # Output: True
```

```python
## maps(), filters() and reduces() can also be used with lambda functions.
#map () applies a function to all items in an input list.
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x ** 2, numbers)) # map uses lazy loading, so we convert the result to a list
print(squared)  # Output: [1, 4, 9, 16, 25]

#map multiple iterables
numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]
summed = list(map(lambda x, y: x + y, numbers1, numbers2))
print(summed)  # Output: [5, 7, 9]

#map with strings, dictionary items
words = ['Ravindra', 'Joisa']
uppercased = list(map(lambda s: s.upper(), words))
print(uppercased)  # Output: ['RAVINDRA', 'JOISA']
```

```python
#maps() conclusion:
# The map() function is a built-in Python function that allows you to apply a specified function to each item in an iterable (like a list, tuple, or set) and returns an iterator with the results.
# It takes two arguments: a function and an iterable.
# The function is applied to each element of the iterable, and the results are collected in an iterator.
# map() is commonly used for data transformation, preprocessing, and applying operations to collections of data.
```

```python
# filter () creates a list of elements for which a function returns true. 
# It will be used to filter out the even numbers from a list based on a condition defined in a lambda function.
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)  # Output: [2, 4, 6]

# filter with lambda function and multiple conditions
mixed_numbers = [10, 15, 20, 25, 30, 35]
filtered_numbers = list(filter(lambda x: x > 15 and x < 30, mixed_numbers))
print(filtered_numbers)  # Output: [20, 25]
```

```python
## Filter() to check if the age is freater than 25 in dictionary
people = [{'name': 'Ravindra', 'age': 40}, {'name': 'Joisa', 'age': 70}, {'name': 'Aniketh', 'age': 7}]
kids = list(filter(lambda person: person['age'] < 25, people))
print(kids)  # Output: [{'name': 'Aniketh', 'age': 7}]
```

```python
# filter function conclusion:
# The filter() function is a built-in Python function that allows you to filter elements from an iterable (like a list, tuple, or set) based on a specified condition.
# It takes two arguments: a function that defines the condition and an iterable to be filtered.
# The function should return True or False for each element in the iterable.
# The filter() function applies the provided function to each element of the iterable and returns an iterator containing only the elements for which the function returned True.
# filter() is used for data cleaning, preprocessing, and extracting relevant information from datasets.
```
