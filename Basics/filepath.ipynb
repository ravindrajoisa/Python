```python
#create a new directory
import os
os.makedirs('new_directory', exist_ok=True) # it will create a new directory named 'new_directory' if it doesn't exist
```

```python
#listing files in a directory
files = os.listdir('.') 
print(files)
```

```python
# Joining paths
file_path = os.path.join('new_directory', 'example.txt') # will join directory and filename if the file is in new_directory.
print(file_path)  # Output: new_directory/example.txt (on Unix-like systems) or
```

```python
# check if the file exists
file_exists = os.path.isfile(file_path)
print(f"Does the file exist? {file_exists}") # new_directory\example.txt (on Windows) 
```

```python
# checking if a path is a directory or a file
is_directory = os.path.isdir('new_directory')
print(f"Is 'new_directory' a directory? {is_directory}")  # Output: True
is_file = os.path.isfile(file_path)
print(f"Is '{file_path}' a file? {is_file}")  # Output: False
# getting absolute path
absolute_path = os.path.abspath(file_path)
print(f"Absolute path: {absolute_path}")
# getting file size
if os.path.isfile(file_path):
    file_size = os.path.getsize(file_path)
    print(f"File size: {file_size} bytes")
else:
    print(f"The file '{file_path}' does not exist, so size cannot be determined.")# Output: new_directory\example.txt (on Windows)
```

```python
#getting absoulute path and relative path
absolute_path = os.path.abspath(file_path)
relative_path = os.path.relpath(absolute_path, start='new_directory')
print(f"Absolute path: {absolute_path}")
print(f"Relative path from 'new_directory': {relative_path}")
```
