```python
# image rotation
```

```python
img_rgb.shape
```

```python
rows, cols = img_rgb.shape[:-1]
rows
```

```python
cols
```

```python
M = cv2. getRotationMatrix2D((rows/2, cols/2), -20, 0.9)  # (find centre of the image by rows/2 and cols/2, rotation degree, zoom in (1.)and zoomout(0.8) or 1 for as is)
```

```python
# wrapAffine() - this function transforms the source image using the specified matri cv2.wrapAffine(img_rgb, M, (cols, rows))
i = cv2.warpAffine(img_rgb, M, (cols, rows))
```

```python
plt.imshow(i)
```

```python
M = np.float32([[1,0,50], [0,1,20]])
```
