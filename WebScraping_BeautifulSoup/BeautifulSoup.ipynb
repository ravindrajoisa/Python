```python
#getting information from all tables
```

```python
data_list = []  # creating a list to hold the dictionaries

for one_table in all_tables:
    all_rows = one_table.find_all('tr')[1:]
    for tr in all_rows:
        all_tds = tr.find_all('td')

        country = all_tds[0].text
        capital = all_tds[1].text
        language = all_tds[-1].text

        ele = {'Country': country, 'Capital': capital, 'Language': language}
#        ele_df = pd.DataFrame(ele)
        data_list.append(ele)       # pappend the dictopnary to the list
        
#        df = df.append(ele, ignore_index = True)
        
        print(country, "==", capital,'==', language)
        print()

# convert the list of dictionaries to a dataframe
df = pd.DataFrame(data_list)
```
