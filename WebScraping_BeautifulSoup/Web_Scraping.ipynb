```python
img_url = "https://www.google.com/imgres?q=lion&imgurl=https%3A%2F%2Fwww.worldanimalprotection.ca%2Fcdn-cgi%2Fimage%2Fwidth%3D1280%2Cformat%3Dauto%2Fsiteassets%2Fshutterstock_2461984615.jpg&imgrefurl=https%3A%2F%2Fwww.worldanimalprotection.ca%2Feducation%2Fanimal-facts%2Flion-facts%2F&docid=EEn53YQJh4brGM&tbnid=YjbIZYMDXmJ2MM&vet=12ahUKEwiry73JseOKAxVjyDgGHcGTKIEQM3oFCIMBEAA..i&w=900&h=506&hcb=2&ved=2ahUKEwiry73JseOKAxVjyDgGHcGTKIEQM3oFCIMBEAA"
```

```python
res = requests.get(img_url)
```

```python
res.status_code
```

```python
binary_data = res.content   # binary data of the image
```

```python
with open("lion.jpg", 'wb') as f:
    f.write(binary_data)
```
